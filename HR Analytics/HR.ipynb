```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Department = LabelEncoder()
```

```python
le_salary = LabelEncoder()
```

```python
df['n_Department'] = le_Department.fit_transform(df['Department'])
df['n_salary'] = le_salary.fit_transform(df['salary'])
```

```python
df
```

```python
df.drop(['Department', 'salary'], axis=1, inplace=True)
```

```python
df
```

```python
df.corr()['left']
```

```python
sns.displot(np.log(df['left']))
```

```python
X=df.drop(columns=['left'])
Y=df['left']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(max_iter=1000,tol=0.00001)
```

```python
model1.fit(x_train, y_train)
```

```python
model1.score(x_test,y_test)
```

```python
y_predicted = model1.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators=50)
```

```python
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
y_predicted = model2.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
